```python
import random
import matplotlib.pyplot as plt
import math
```

```python
a = 0
b = 0
choices = [-1, 1]

def nCr(n,r):
    f = math.factorial
    return f(n) / f(r) / f(n-r)


sim = []
ana = []

for i in range(100):
    sim.append(0)
    ana.append(nCr(2*i, i)/4**i)
    
    for j in range(1000):
        a = 0
        b = 0
        for k in range(i):
            a += random.choice(choices)
            b += random.choice(choices)
        if a == b:
            sim[i] += 1
    sim[i] /= 1000



x = []
for i in range(100):
    x.append(i)

plt.figure(figsize=(20,10))
plt.title('Binary Drunk Walk')
plt.xlabel('Number of steps')
plt.ylabel('Probability of meeting after X steps')
plt.plot(x, sim)
plt.plot(x, ana)
```

```python

pos = 0

lis = []
lis1 = []
lis2 = []
mean = []
squared = []

for i in range(100):
    lis.append(0)
    lis1.append(0)
    lis2.append(0)
    mean.append(0)
    squared.append(0)
    for j in range(1000):
        pos = 0
        for k in range(i):
            pos += random.choice(choices)
        if pos == 0:
            lis[i] += 1
        if pos == 5:
            lis1[i] += 1
        if pos == -5:
            lis2[i] += 1
        mean[i] += pos
        squared[i] += pos**2
        
    lis[i] /= 1000
    lis1[i] /= 1000
    lis2[i] /= 1000
    mean[i] /= 1000
    squared[i] /= 1000

x = []
for i in range(100):
    x.append(i)


plt.figure(figsize=(20,10))
plt.title('Mono Drunk Walk')
plt.xlabel('Number of steps')
plt.ylabel('Probability of reaching location after X steps')
plt.plot(x, lis, label="Destination at X = 0")
#plt.plot(x, lis1, label="Destination at X = 5")
#plt.plot(x, lis2, label="Destination at X = -5")
plt.legend()
```

```python
plt.figure(figsize=(10,10))
plt.plot(x, x, label="True mean squared displacement")
plt.plot(x, squared, label="Simulated mean squared displacement")
plt.legend()
```
