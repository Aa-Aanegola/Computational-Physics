```python
import matplotlib.pyplot as plt
```

```python
def sim(N1, N2, alpha, beta, r, c, k):
    arrN1 = [N1]
    arrN2 = [N2]
    arrT = [0]
    t = 0
    for i in range(50000):
        dN1 = dt*(r*N1*(1-N1/k) - alpha*N1*N2)
        dN2 = dt*(-c*N2+beta*N1*N2)
        N1 += dN1
        N2 += dN2
        t += dt
        arrN1.append(N1)
        arrN2.append(N2)
        arrT.append(t)
    return arrN1, arrN2, arrT
```

```python
N1, N2, alpha, beta, r, c, k = 22, 1, 0.3, 0.2, 1, 0.3, 25
arrN1, arrN2, arrT = sim(N1, N2, alpha, beta, r, c, k)
print(f"N1:{N1} N2:{N2} alpha:{alpha} beta:{beta} r:{r} c:{c} k:{k}")
plt.plot(arrT, arrN1)
plt.plot(arrT, arrN2)
plt.legend(["Prey", "Predator"])
plt.xlabel("Time")
plt.ylabel("Population size")
plt.show()
```
