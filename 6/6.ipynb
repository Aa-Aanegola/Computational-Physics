```python
import random
import matplotlib.pyplot as plt
```

```python
x_space = [0, 1]
y_space = [3*i**2 for i in x_space]
```

```python
integral = []

for i in range(0, 10000, 10):
    count = 0
    for j in range(i):
        x = random.uniform(x_space[0], x_space[1])
        y = random.uniform(y_space[0], y_space[1])
        if y <= 3*x**2:
            count += 1
    if i:
        count /= i
        count *= (x_space[1] - x_space[0]) * (y_space[1] - y_space[0])
    integral.append(count)
x_array = range(0, 10000, 10)
plt.plot(x_array, integral)
plt.xlabel("Number of points")
plt.ylabel("Integral value")
```

```python
integral_fixed20 = []
x_range = []
for i in range(100):
    count = 0
    x_range.append(i)
    for j in range(20):
        x = random.uniform(x_space[0], x_space[1])
        y = random.uniform(y_space[0], y_space[1])
        if y <= 3*x**2:
            count += 1
    
    count /= 20
    count *= (x_space[1] - x_space[0]) * (y_space[1] - y_space[0])
    integral_fixed20.append(count)
    
mean20 = sum(integral_fixed20)/100
variance20 = 0
for i in integral_fixed20:
    variance20 += (i-mean20)**2
variance20 /= 100
print("Standard deviation", variance20**0.5)
plt.plot(x_range, integral_fixed20)
plt.legend(["Fixed N=20"])
plt.xlabel("Trial")
plt.ylabel("Integral value")
plt.show()
```

```python
integral_fixed1000 = []
for i in range(100):
    count = 0
    for j in range(1000):
        x = random.uniform(x_space[0], x_space[1])
        y = random.uniform(y_space[0], y_space[1])
        if y <= 3*x**2:
            count += 1
    
    count /= 10000
    count *= (x_space[1] - x_space[0]) * (y_space[1] - y_space[0])
    integral_fixed1000.append(count)
    
mean1000 = sum(integral_fixed1000)/100
variance1000 = 0
for i in integral_fixed1000:
    variance1000 += (i-mean1000)**2
variance1000 /= 100
print("Standard deviation: ", variance1000**0.5)
plt.plot(x_range, integral_fixed1000)
plt.legend(["Fixed N=1000"])
plt.xlabel("Trial")
plt.ylabel("Integral value")
plt.show()
```

```python
nVals = [20, 100, 500, 1000, 2000, 4000, 5000]
stdVals = []
for n in nVals:
    integral = []
    for j in range(100):
        count = 0
        for i in range(n):  
            x = random.uniform(x_space[0], x_space[1])
            y = random.uniform(y_space[0], y_space[1])
            if y < 3*x**2:
                count += 1
        count /= n
        count *= (x_space[1] - x_space[0]) * (y_space[1] - y_space[0])
        integral.append(count)
    mean = sum(integral)/100
    variance = 0
    for inte in integral:
        variance += (inte-mean)**2
    variance /= 100
    stdVals.append(variance**0.5)
plt.scatter(nVals, stdVals)
plt.xlabel("Number of points")
plt.ylabel("Standard deviation")
plt.show()
```
