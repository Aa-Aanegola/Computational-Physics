```python
import math
import matplotlib.pyplot as plt 
import numpy as np
```

```python
m = 0.1
k = 0.1
legend = []
phaseX = []
phaseT = []
allX = []
c = 0
for H in np.arange(0, 1, 0.1):
    allX.append([])
    xList = []
    pList = []
    tList = []
    x = 0
    p = math.sqrt(2*m*H)
    t = 0
    dt = 0.001
    for i in range(0, 10000):
        tList.append(t)
        pList.append(p)
        xList.append(x)
        allX[c].append(x)
        dp = -k*x*dt
        dx = dt*p/m
        p += dp
        x += dx
        t += dt
    c += 1
    legend.append(f"H={H:.3}")
    plt.plot(xList, pList)
    
    phaseX = xList
    phaseT = tList
plt.xlabel("x")
plt.ylabel("p")
plt.legend(legend)
plt.show()
```

```python
msd = []
for i in range(len(phaseX)):
    msd.append(phaseX[i]**2)

plt.plot(phaseT, msd)
plt.title("Mean squared distance")
plt.xlabel("t")
plt.ylabel("msd")
plt.show()
```
