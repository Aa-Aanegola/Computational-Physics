```python
import random
import matplotlib.pyplot as plt
import math
```

```python
lis = []

for i in range(1, 5000):
    pi = 0
    for j in range(i):
        x = random.random()
        y = random.random()
        dist = (x**2+y**2)**0.5
        if dist <= 1:
            pi += 4
    pi /= i
    lis.append(pi)

x = []
for i in range(1, 5000):
    x.append(i)


plt.figure(figsize=(20,10))
plt.title('Estimating pi')
plt.xlabel('Number of iterations')
plt.ylabel('Value of pi')
plt.plot(x, lis)
```
